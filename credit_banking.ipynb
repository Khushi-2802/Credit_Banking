```python
 Customer_expenditure['Month']=pd.to_datetime(Customer_expenditure['Month'])
```

```python
Customer_expenditure['m_on_th']=Customer_expenditure['Month'].dt.month_name()
```

```python
Customer_expenditure['year']=Customer_expenditure['Month'].dt.year
```

```python
Customer_expenditure
```

```python
Customer_Expense=Customer_expenditure.groupby(['Customer','m_on_th','year']).agg({'Amount': 'sum'}).reset_index()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
Customer_Expense
```

```python
Customer_repayment['Month']=pd.to_datetime(Customer_repayment['Month'])
Customer_repayment['m_on_th']=Customer_repayment['Month'].dt.month_name()
Customer_repayment['year']=Customer_repayment['Month'].dt.year
Customer_Repay=Customer_repayment.groupby(['Customer','m_on_th','year']).agg({'Amount': 'sum'}).reset_index()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
Customer_Repay
```

```python
Customer_Repay.rename(columns={'Amount':'Amount_repayed'},inplace=True)
Customer_Expense.rename(columns={'Amount':'Amount_spent'},inplace=True)
```

```python
#merging based on customer month year
merged_data = pd.merge(Customer_Expense,Customer_Repay, on=['Customer','m_on_th','year'])
merged_data
```

```python
#Calculate surplus amount
merged_data['credit'] = (merged_data['Amount_repayed'] - merged_data['Amount_spent'])*0.02
#df.loc[df['Values'] < 0, 'Values'] = 0
merged_data.loc[merged_data['credit']<0,'credit']=0
merged_data
merged_data.to_csv(r'C:\Users\PRAVEEN CHANDRA\Desktop\credit.csv')
```

```python
surplus_customer_table= merged_data[merged_data['Amount_repayed'] >merged_data['Amount_spent']]

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

next_month_billing = pd.DataFrame(columns=['customer', 'month', 'year', 'billing amount'])
for index, row in surplus_customer_table.iterrows():
    current_month_index = month_order.index(row['m_on_th'])
    next_month_index = (current_month_index + 1) % len(month_order)
    next_month = month_order[next_month_index]
    next_year = row['year'] if next_month_index >= current_month_index else row['year'] + 1
    next_month_billing = next_month_billing.append({
        'customer': row['Customer'],
        'month': next_month,
        'year': next_year,
        'billing amount': row['credit']
    }, ignore_index=True)

next_month_billing.to_csv(r'C:\Users\PRAVEEN CHANDRA\Desktop\credit_in_next_month_billing.csv')
```

```python
#joining merged_data table with customer details
Final_table= pd.merge(merged_data, Customer_Acquisition[['Customer', 'Age', 'Segment']], on='Customer')
Final_table
```

```python
#people in which segment are spending more money
Highest_paying_segment = Final_table.groupby('Segment')['Amount_spent'].sum()

# Sorting the segments based on the total spending in descending order
Highest_paying_segment = Highest_paying_segment.sort_values(ascending=False)
Highest_paying_segment.to_csv(r'C:\Users\PRAVEEN CHANDRA\Desktop\highest_paying_segment.csv')
```

```python
#which age group is spending more money
A_bins=[18,35,50,70,90]
A_labels=['18-35','36-50','50-70','70-90']
Final_table['age_group'] = pd.cut(Final_table['Age'], bins=A_bins, labels=A_labels, right=False)
AGE_GROUP_SPENDING=Final_table.groupby('age_group')['Amount_spent'].sum()
AGE_GROUP_SPENDING=AGE_GROUP_SPENDING.sort_values(ascending=False)
AGE_GROUP_SPENDING.to_csv(r'C:\Users\PRAVEEN CHANDRA\Desktop\age_group_spending.csv')
```
